```python
def initilize():
    for i in range(len(All_Stations)):
        All_Stations[i][2]=0
```

```python
def Judgement(tmpName="", currentLineNo=-1):
    tmpChangeLine=-1
    for i in range(len(All_Stations)):
        if All_Stations[i][0]==tmpName:
            if currentLineNo in All_Stations[i][1]:
                if All_Stations[i][2]==0: #Succeed
                    All_Stations[i][2]=1 #Change to Used
                    print("성공하였습니다")
                    if len(All_Stations[i][1]) > 1: #환승역
                        try:
                            tmpChangeLine=int(input("환승하시겠습니까? 1-예 0-아니오 : "))
                        except:
                            #잘못된 응답으로 탈락
                            return [0, currentLineNo]
                        
                        if tmpChangeLine==1:
                            print("** 호선 정보 **")
                            print("1 : 서울 1호선, 2 : 서울 2호선, 3 : 서울 3호선, 4 : 서울 4호선, 5 : 서울 5호선, 6 : 서울 6호선, 7 : 서울 7호선, 8 : 서울 8호선,")
                            print("9 : 서울 9호선, 10 : 인천 1호선, 11 : 인천 2호선, 12 : 공항철도, 13 : 경의.중앙선, 14 : 수인.분당선, 15 : 우이신설선, 16 : 신분당선,")
                            print("17 : 에버라인, 18 : 의정부경전철, 19 : 김포골드라인, 20 : 신림선, 21 : 경춘선, 22 : 경강선, 23 : 서해선, 24 : GTX-A")
                            tmpChangeLine=-1
                            try:
                                tmpChangeLine=int(input("어떤 호선으로 환승할까요? : "))
                            except:
                                #잘못된 응답으로 탈락
                                return [0, currentLineNo]
                            
                            if tmpChangeLine in All_Stations[i][1] and tmpChangeLine != currentLineNo:
                                #환승 성공
                                return [1, tmpChangeLine]
                            else:
                                #잘못된 응답으로 탈락
                                return [0, currentLineNo]
                        elif tmpChangeLine==0:
                            #환승 안함
                            return [1, currentLineNo]
                        else:
                            #잘못된 응답으로 탈락
                            return [0, currentLineNo]
                    else:
                        #환승역 아님
                        return [1, currentLineNo]
                else:
                    #print("중복")
                    continue
            else:
                #print("해당 노선 역이 아님")
                continue
        else:
            #print("등록되지 않은 역") # 다른 역 검색시에도 등장함
            continue
        
    return [0, currentLineNo]
```
